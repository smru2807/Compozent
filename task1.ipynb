```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
df = df.rename(columns={"creatinine_phosphokinase": "CP", "ejection_fraction": "EFraction", "high_blood_pressure": "HighBP", "DEATH_EVENT": "Death" })
df.head(8)
```

```python
duplicate_rows_data = df[df.duplicated()]
print("Number of Duplicate rows: ", duplicate_rows_data.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(8)
```

```python
df['HighBP'].describe()
```

```python
sns.boxplot(x=df['anaemia'], color="pink")
```

```python
sns.boxplot(x=df['age'], color="pink")
```

```python
sns.boxplot(x=df['EFraction'], color="pink")
```

```python
plt.title('HIGH BLOOD PRESSURE DISTRIBUTION')
highbp_counts = df['HighBP'].value_counts()
plt.pie(highbp_counts, labels=highbp_counts.index, autopct='%1.2f%%', colors = ["pink", "yellow"])
plt.show()
```

```python
df['diabetes'].value_counts()
```

```python
plt.title('SMOKING')
round(df['smoking'].value_counts()/df.shape[0]*100,2).plot.pie(autopct = "%1.0f%%",colors = ["skyblue" , "orange" ])
```

```python
sns.barplot(df ,x ="EFraction" ,y="age",color='pink')
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['age'], df['serum_creatinine'], color='brown')
ax.set_xlabel('age')
ax.set_ylabel('serum_creatinine')
plt.show()
```
